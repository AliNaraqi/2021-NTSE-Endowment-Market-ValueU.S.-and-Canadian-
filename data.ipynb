```python
data=data.rename(columns=rename_map)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.float_format = '{:,.0f}'.format


required_cols = ['FY21_Endowment_$k', 'FY20_Endowment_$k', 'Pct_Change', 'FTE_2020', 'FY21_Per_FTE_$', 'IPEDS_Sector', 'Carnegie', 'State', 'Institution Name']
missing = [c for c in required_cols if c not in data.columns]
if missing:
    raise KeyError(f"Missing expected columns after rename: {missing}")

num_cols = ['FY21_Endowment_$k', 'FY20_Endowment_$k', 'Pct_Change', 'FTE_2020', 'FY21_Per_FTE_$']
for c in num_cols:
    cleaned = (
        data[c]
        .astype(str)
        .str.replace(r"[^0-9.\-]", "", regex=True)
        .replace({'': np.nan, '-': np.nan})
    )
    data[c] = pd.to_numeric(cleaned, errors='coerce')


data['FY21_Endowment_$'] = data['FY21_Endowment_$k'] * 1000.0
data['FY20_Endowment_$'] = data['FY20_Endowment_$k'] * 1000.0

# Derive simple Type from IPEDS sector
sector = data['IPEDS_Sector'].astype(str)
data['Type'] = pd.NA
data.loc[sector.str.contains('Public', case=False, na=False), 'Type'] = 'Public'
data.loc[sector.str.contains('Private', case=False, na=False), 'Type'] = 'Private'

canadian_provinces = {
    'AB','BC','MB','NB','NL','NS','NT','NU','ON','PE','QC','SK','YT'
}
state_str = data['State'].astype(str).str.strip()
data['Country'] = np.where(state_str.isin(canadian_provinces), 'Canada', 'U.S.')

# Compute % change if missing
if data['Pct_Change'].isna().any():
    with np.errstate(divide='ignore', invalid='ignore'):
        calc_pct = (data['FY21_Endowment_$'] - data['FY20_Endowment_$']) / data['FY20_Endowment_$'] * 100.0
    data['Pct_Change'] = data['Pct_Change'].fillna(calc_pct)

# Ensure per-FTE present; compute if missing and FTE available
needs_per_fte = data['FY21_Per_FTE_$'].isna() & data['FTE_2020'].notna() & (data['FTE_2020'] > 0)
data.loc[needs_per_fte, 'FY21_Per_FTE_$'] = data.loc[needs_per_fte, 'FY21_Endowment_$'] / data.loc[needs_per_fte, 'FTE_2020']

# Keep a clean working frame
cols_keep = [
    'Institution Name','State','Country','Type','Carnegie',
    'FTE_2020','FY21_Endowment_$','FY20_Endowment_$','FY21_Per_FTE_$','Pct_Change'
]
df = data[cols_keep].copy()

df.head()
```

```python
# Wealth concentration analyses: State/Province, Carnegie shares, Top-10 share

# States/Provinces holding most total wealth
state_totals = (
    df.groupby(['Country','State'], dropna=False)['FY21_Endowment_$']
      .sum()
      .reset_index()
      .rename(columns={'FY21_Endowment_$':'total_endowment_usd'})
      .sort_values('total_endowment_usd', ascending=False)
)
print('Top 20 States/Provinces by total FY21 endowment:')
display(state_totals.head(20))

print('\nTop 10 U.S. states:')
display(state_totals[state_totals['Country']=='U.S.'].head(10))
print('\nTop 10 Canadian provinces:')
display(state_totals[state_totals['Country']=='Canada'].head(10))

# Carnegie categories dominance and shares
car_totals = (
    df.groupby('Carnegie', dropna=False)['FY21_Endowment_$']
      .sum()
      .reset_index()
      .rename(columns={'FY21_Endowment_$':'total_endowment_usd'})
      .sort_values('total_endowment_usd', ascending=False)
)
car_totals['share_pct'] = 100.0 * car_totals['total_endowment_usd'] / car_totals['total_endowment_usd'].sum()
print('\nCarnegie categories by total endowment and shares:')
display(car_totals)

# 3) Concentration: Top 10 schools vs rest
total_all = df['FY21_Endowment_$'].sum(skipna=True)
by_inst = df[['Institution Name','FY21_Endowment_$']].dropna().groupby('Institution Name', as_index=False).sum()
by_inst = by_inst.sort_values('FY21_Endowment_$', ascending=False)

top10_sum = by_inst.head(10)['FY21_Endowment_$'].sum()
rest_sum = total_all - top10_sum
share_top10 = 100.0 * top10_sum / total_all if total_all else np.nan
share_rest = 100.0 - share_top10 if pd.notna(share_top10) else np.nan

print('\nConcentration of total FY21 endowment:')
print(f"Top 10 institutions share: {share_top10:.2f}%  (Top10=${top10_sum:,.0f} of Total=${total_all:,.0f})")
print(f"Rest share: {share_rest:.2f}%  (Rest=${rest_sum:,.0f})")

print('\nTop 10 institutions by total FY21 endowment:')
display(by_inst.head(10))
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.impute import SimpleImputer

model_df = df.dropna(subset=['Pct_Change','FY20_Endowment_$']).copy()
model_df['log_FY20'] = np.log1p(model_df['FY20_Endowment_$'])
model_df['log_FTE'] = np.log1p(model_df['FTE_2020'].fillna(0))
model_df['perFTE'] = model_df['FY21_Per_FTE_$']
# Clean categorical columns to avoid NAType in encoder
model_df['Type_str'] = model_df['Type'].astype('string').fillna('Unknown')
model_df['Country_str'] = model_df['Country'].astype('string').fillna('Unknown')
model_df['Carnegie_str'] = model_df['Carnegie'].astype('string').fillna('Unknown')

X = model_df[['log_FY20','log_FTE','perFTE','Type_str','Country_str','Carnegie_str']]
y = model_df['Pct_Change']

num_features = ['log_FY20','log_FTE','perFTE']
cat_features = ['Type_str','Country_str','Carnegie_str']

preprocess = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())]), num_features),
        ('cat', Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')), ('onehot', OneHotEncoder(handle_unknown='ignore'))]), cat_features),
    ]
)

pipe = Pipeline([
    ('prep', preprocess),
    ('model', Ridge(alpha=1.0))
])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
pipe.fit(X_train, y_train)

pred = pipe.predict(X_test)
print(f"R^2 (test): {r2_score(y_test, pred):.3f}")
print(f"MAE (test): {mean_absolute_error(y_test, pred):.2f} pct-pts")

cv_r2 = cross_val_score(pipe, X, y, cv=5, scoring='r2')
print(f"CV R^2 (mean±sd): {cv_r2.mean():.3f} ± {cv_r2.std():.3f}")
```

```python
# Clustering: size, growth, and per-student wealth (K-Means)
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

clu_df = df[['Institution Name','Country','Type','FTE_2020','FY21_Endowment_$','FY21_Per_FTE_$','Pct_Change']].copy()
clu_df = clu_df.dropna(subset=['FTE_2020','FY21_Endowment_$','FY21_Per_FTE_$','Pct_Change'])

features = pd.DataFrame({
    'log_FTE': np.log1p(clu_df['FTE_2020']),
    'log_Endowment': np.log1p(clu_df['FY21_Endowment_$']),
    'PerFTE': clu_df['FY21_Per_FTE_$'],
    'PctChange': clu_df['Pct_Change'],
})

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

k = 4
kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
labels = kmeans.fit_predict(X_scaled)
clu_df['Cluster'] = labels
sil = silhouette_score(X_scaled, labels)
print(f'Silhouette score (k={k}): {sil:.3f}')

# Visualize clusters (2D projections)
plt.figure(figsize=(8,6))
ax = sns.scatterplot(
    x=features['log_Endowment'], y=features['log_FTE'], hue=clu_df['Cluster'].astype(str),
    style=clu_df['Country'], alpha=0.7, palette='tab10'
)
ax.set_title('Clusters by Size (log Endowment) vs Size (log FTE)')
ax.set_xlabel('log(FY21 Endowment $)')
ax.set_ylabel('log(FTE)')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,6))
ax = sns.scatterplot(
    x=features['PerFTE'], y=features['PctChange'], hue=clu_df['Cluster'].astype(str),
    style=clu_df['Country'], alpha=0.7, palette='tab10'
)
ax.set_title('Clusters by Per-FTE Wealth vs % Change')
ax.set_xlabel('FY21 Endowment per FTE ($)')
ax.set_ylabel('% Change FY20→FY21')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Show sample per cluster
sample_per_cluster = clu_df.groupby('Cluster').apply(lambda g: g.head(5)).reset_index(drop=True)
display(sample_per_cluster[['Institution Name','Country','Type','Cluster','FTE_2020','FY21_Endowment_$','FY21_Per_FTE_$','Pct_Change']])
```
